```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10,1000)
y1=np.sqrt((np.exp(x) + np.cosh(x))*(x/3))
y2= (0.5 *x)/(2+ np.sinh(x))
plt.figure(figsize=(8,4))

plt.plot(x,y1,'--r',lw=2,label='f(x)')
plt.plot(x,y2,'-g',lw=2,label='g(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plotting two curves')
plt.legend()
plt.show()
```

```python
student_ids = ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9', 'S10']  
maths_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]  
physics_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]  
```

```python
plt.figure(figsize=(8, 3))  
plt.scatter(student_ids, maths_marks, color='red', label='Maths', alpha=0.6, s=100)  
plt.scatter(student_ids, physics_marks, color='blue', label='Physics', alpha=0.6)  
plt.xlabel('Student ID')  
plt.ylabel('Marks in Maths & Physics')  
plt.title('Comparison of Marks in Maths and Physics Subjects')  
plt.legend()  
plt.show()
```

```python
plt.figure(figsize=(8, 3))  
x = np.arange(len(student_ids))  
width = 0.35  
plt.bar(x - width/2, maths_marks, width, label='Maths', color='blue', alpha=0.4)  
plt.bar(x + width/2, physics_marks, width, label='Physics', color='green', alpha=0.4)  
plt.xlabel('ID')  
plt.ylabel('Grades in Maths & Physics')  
plt.title('Comparison of Marks in Maths and Physics Subjects')  
plt.xticks(x, student_ids)  
plt.legend()  
plt.show()  
```
