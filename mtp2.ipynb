```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-5,2,100)
y1=x**3+x**2+10
y2=3*x**2+6*x
y3=x-10
fig=plt.figure(figsize=(8,5))
plt.plot(x,y1,label='cubic')
plt.plot(x,y2,label='quadratic')
plt.plot(x,y3,label='linear')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plots')
plt.legend()
plt.show()
```

```python
x=np.linspace(10,20,100)
f=-20*np.cos(x)*np.sin(x)
g=4*np.sin(x)-np.cos(x)
plt.figure(figsize=(18,5))
plt.plot(x,f,'-b',label='$f(x)=-20cos(x)sin(x)$')
plt.plot(x,g,'--g',label='$g(x)=4sin(x)-cos(x)$')
plt.xlabel('X-Data')
plt.ylabel('Plots of f(x) & g(x) ')
plt.fill_between(x,f,g,where=(f<g),alpha=0.2,color='magenta')
plt.title('PLOTTING EXAMPLE')
plt.legend(loc=8)
```

```python
x=np.linspace(-2,2,200)
f1=np.exp(-x**2)
f2=-np.exp(-x**2)
g=np.cos(40*x)
plt.figure(figsize=(10,5))
plt.plot(x,f1,'-b',lw=2,label='$f(x)=exp(-x^2)$')
plt.plot(x,f2,'-b',lw=2,label='$-f(x)=-exp(-x^2)$')
plt.plot(x,g,'-.g',lw=2,label='$g(x)=cos(40x)$')
plt.fill_between(x,f1,f2,alpha=0.4,color='yellow')
plt.xlabel('x-data')
plt.ylabel('Plots of f(x),-f(x) and g(x)')
plt.title('plots')
plt.legend()
```

```python
x=np.random.randint(1,5,size=(10))
labels=["Data " +str(i+1) for i in range(x.size)]
fig=plt.figure(figsize=(8,3))
plt.pie(x,labels=labels,labeldistance=1.15,rotatelabels=True)
plt.show()
```
